### Data Cleaning

```python
isp.rename(columns={'boro': 'borough'}, inplace= True)
```

```python
isp.rename(columns={'camis': 'record_id','record_date': 'date_of_data_pull', }, inplace=True)
```

```python
#Check for and drop any duplicate rows
print('length before drop:', len(isp))

isp = isp.drop_duplicates()

print('length after drop:', len(isp))
```

* 6 duplicates 


```python
isp['inspection_date'] = pd.to_datetime(isp['inspection_date'], format = '%m/%d/%Y')
isp['inspection_date'].sort_values(ascending = False)

dates = pd.DataFrame(isp['inspection_date'].value_counts())
dates.reset_index(inplace = True)
dates = dates.sort_values(by = 'index', ascending = False)
dates = dates.loc[(dates['index'] > '1900-01-01')]
dates
```

```python
plt.plot(dates['index'], dates['inspection_date'])
plt.xlabel("Date of inspection")
plt.ylabel("Number of inspections")
plt.title('Inspection date by frequency')
```

Interesting note: The inspection frequency anomalies in 2020 through mid-2021. In 2020, the DOH paused inspections due to the COVID-19 pandemic and ensuing shutdowns. Inspection frequencies peaked after the pause was lifted as restaurants that were still open were all due for inspections, and new or re-opening restaurants were due for initial inspections.
